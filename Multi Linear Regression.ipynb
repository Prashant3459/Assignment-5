```python
# Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_RDS = smf.ols('RDS~ADM+MKTS', data = data).fit().rsquared
vif_RDS = 1/(1-rsq_RDS)

rsq_ADM = smf.ols('ADM~RDS+MKTS', data = data).fit().rsquared
vif_ADM = 1/(1-rsq_ADM)

rsq_MKTS = smf.ols('MKTS~RDS+ADM', data = data).fit().rsquared
vif_MKTS = 1/(1-rsq_MKTS)
```

```python
SD = {'Variable':['RDS','ADM','MKTS'], 'VIF':[vif_RDS, vif_ADM, vif_MKTS]}
Vif_frame = pd.DataFrame(SD)
Vif_frame
```

```python
# Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
(c,_) = ml.get_influence().cooks_distance
c
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k = data.shape[1]
n = data.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
data1 = data.drop(data.index[[49]],axis = 0).reset_index(drop = True)
data1
```

```python
# Model Deletion Diagnostics and Final Model

while np.max(c)>0.5 :
    ml = smf.ols('Profit~RDS+ADM+MKTS',data = data1).fit()
    (c,_) = ml.get_influence().cooks_distance
    c
    np.argmax(c), np.max(c)
    data1 = data1.drop(data1.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    data1
else:
    final_model = smf.ols("Profit~RDS+ADM+MKTS", data = data1).fit()
    final_model.rsquared, final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```
